```python
X = df.drop('SalePrice', axis=1)
y = np.ravel(np.array(df[['SalePrice']]))
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def rmse(y, y_pred):
    return np.sqrt(mean_squared_error(np.log(y), np.log(y_pred)))
```

```python
# Initialize the model
random_forest = RandomForestRegressor(n_estimators=1200,
                                      max_depth=15,
                                      min_samples_split=5,
                                      min_samples_leaf=5,
                                      max_features=None,
                                      random_state=42,
                                      oob_score=True
                                     )

# Perform cross-validation to see how well our model does 
kf = KFold(n_splits=5)
y_pred = cross_val_score(random_forest, X, y, cv=kf, n_jobs=-1)
y_pred.mean()
```

```python
# Fit the model to our data
random_forest.fit(X, y)
```

```python
# Make predictions on test data
rf_pred = random_forest.predict(X_test)
```

```python
print("RMSE of RandomForestRegressor: {}".format(rmse(y_test,rf_pred)))
```

```python
# Initialize our model
g_boost = GradientBoostingRegressor( n_estimators=6000, learning_rate=0.01,
                                     max_depth=5, max_features='sqrt',
                                     min_samples_leaf=15, min_samples_split=10,
                                     loss='ls', random_state =42
                                   )

# Perform cross-validation to see how well our model does 
kf = KFold(n_splits=5)
y_pred = cross_val_score(g_boost, X, y, cv=kf, n_jobs=-1)
y_pred.mean()
```

```python
# Fit our model to the training data
g_boost.fit(X,y)
```

```python
# Make predictions on test data
gbm_pred = g_boost.predict(X_test)
```

```python
print("RMSE of GradientBoostingRegressor: {}".format(rmse(y_test,gbm_pred)))
```

```python

# Initialize our model
linear_reg = LinearRegression()
                                   

# Perform cross-validation to see how well our model does 
kf = KFold(n_splits=5)
y_pred = cross_val_score(linear_reg, X, y, cv=kf, n_jobs=-1)
y_pred.mean()
```

```python
linear_reg.fit(X,y)
```

```python
lr_pred = linear_reg.predict(X_test)
```

```python
print("RMSE of Linear Regression: {}".format(rmse(y_test,lr_pred)))
```

```python
# Initialize our model
las = Lasso()

# Perform cross-validation to see how well our model does 
kf = KFold(n_splits=5)
y_pred = cross_val_score(las, X, y, cv=kf, n_jobs=-1)
y_pred.mean()
```

```python
las.fit(X,y)
```

```python
las_pred = las.predict(X_test)
```

```python
print("RMSE of Lasso: {}".format(rmse(y_test,las_pred)))
```

Since GradientBoostingRegressor has the least RMSE value, it is the best fit
